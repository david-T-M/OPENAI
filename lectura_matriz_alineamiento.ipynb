```python
def getRelacionesMA(indice,df):
    todas_relaciones=[]
    t=pd.DataFrame(df.at[indice,"Text"][1:])
    h=pd.DataFrame(df.at[indice,"Hipotesis"][1:])
    ma=pd.DataFrame(df.at[indice,"M_Align"][1:len(t)+1,1:len(h)+1],columns=h,index=t)
    for col in ma.columns:
        indices=[]
        if 'NOUN' in str(col) or 'VERB' in str(col) or 'ADJ' in str(col) or 'PRON' in str(col):
            qu75=ma[col].quantile(0.75)
            #print(ma[col].idxmax,ma[col].max)
            for index,strings in ma.iterrows():
                #print(index)
                if 'NOUN' in str(index) or 'VERB' in str(index):
                    if strings[col]>=qu75:
                        #print(index,strings)
                        indices.append((index,col,strings[col]))
                        todas_relaciones.append((index,col))
    for index,strings in ma.iterrows():    
        if 'NOUN' in str(index) or 'VERB' in str(index) or 'PRON' in str(index):
            qu75=strings.quantile(0.75)
            #print(qu75)
            indices=[]
            for col in ma.columns:
                if 'NOUN' in str(col) or 'VERB' in str(col) or 'PRON' in str(col):
                    #print(qu75)
                    if strings[col]>=qu75:
                        #print(index,col)
                        indices.append((index,col,strings[col]))
                        todas_relaciones.append((index,col))
    return set(todas_relaciones)
```

```python
import nltk, re, os
from nltk.corpus import wordnet as wn
import networkx as nx
from matplotlib import pyplot as plt
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")
```

```python
def tokenizar(texto,nlp,wn,hipotesis=False):
    sto =['a','and','',' ',None,'A','to','of','or']
    sinonimos=set()
    #print(texto)
    lista=[]
    doc = nlp(texto)
    l=[]
    for token in doc:
        #print(token.text, token.dep_, token.head.text, token.head.pos_)
        #print(str(token.pos_))
        if str(token.pos_) in ['NOUN','ADV','ADJ','VERB','PRON']:
            #print(token.text,token.pos_)
            sinonimos=set()
            if(token.pos_ == 'NOUN' or token.pos_ == 'PRON'):
                w =wn.synsets(token.text,wn.NOUN)                
            elif(token.pos_ == 'VERB'):
                w =wn.synsets(token.text,wn.VERB)
            elif(token.pos_ == 'ADJ'):
                w =wn.synsets(token.text,wn.ADJ)
            elif(token.pos_ == 'ADV'):
                w =wn.synsets(token.text,wn.ADV)
            
            #SINONIMOS
            if w==[]:
                w = w =wn.synsets(token.lemma_)
            for wi in w[:2]:
                #print(wi)
                for sin in wi.lemma_names():
                    sinonimos.add(sin)
                    #print("Checar ",sin)
                    #print("definition:",wi,wi.definition())
                    d_ = nlp(wi.definition())
                    for t in d_:
                        for child in t.children:
                            if child.text not in sto and t.text not in sto:
                                #lista.append((str(t.text)+","+str(t.pos_),str(child.text)+","+str(child.pos_)))
                                #print((str(t.text)+","+str(t.pos_),str(child.text)+","+str(child.pos_)))
                                #if child.pos_ in ['NOUN','ADJ']:
                                    #lista.append((str(sin)+","+str(token.pos_),str(child.text)+","+str(child.pos_)))
                                lista.append((str(sin)+","+str(token.pos_),str(child.text)+",A"))

            for e in sinonimos:
                lista.append((e+",S",str(token.text)+","+str(token.pos_)))
                #lista.append((str(token.text)+",S",e))
            if hipotesis==False:
                # HIPERONIMOS
                for wi in w[:2]:
                    paths = wi.hypernym_paths()
                    #print("CAMINOS",wi,paths)
                    for p in paths:
                        if(len(p)>1):
                            #print("caminos",p)
                            j=0
                            for pi in p[::-1]:
                                #print(pi,token.text)
                                #print(pi.lemma_names())
                                for pis in pi.lemma_names():
                                    lista.append((pis+",H",str(token.text)+","+str(token.pos_)))
                                    #lista.append((str(token.text)+",H",pis))
                                #if j==0:
                                #    primero=pi
                                #    lista.append((str(token.text)+","+str(token.head.pos_),pi))
                                #    j+=1
                                #else:
                                #    print("nodo",pi,pi.definition())
                                #    lista.append((primero,pi))
                                #    primero=pi
                #for p in paths:
                        #    for pi in p:
                        #        print(pi,pi.lemma_names(),pi.definition())
                        #for li in wi.lemma_names():
                        #    lista.append((token.text,li))
                #print(sinonimos)
            
            for child in token.children:
                if child.text not in sto:
                    lista.append((str(token.text)+","+str(token.pos_),str(child.text)+","+str(child.pos_)))
            l=set(lista)
    return list(l)
```

```python
def revision(relaciones_alineadas,wn,GH,G):
    faltantes = []
    existentes=[]
    relacion = []
    razonamiento=[]
    palabras_encontradas=[]
    restantes_rel_ali=relaciones_alineadas[:]
    for r_a in relaciones_alineadas:
        max_r=0
        max_root=0
        max_lch=0
        cam_sh=0
        print(r_a[0],"a",r_a[1])
        word_t = str(r_a[0]).split('\'')[1]
        wt = word_t.split('{')[0]
        lt = word_t.split('}')[0].split(',')[0].split('{')[1]
        pt = word_t.split('}')[0].split(',')[1]
        w_in_text= wt + "," + pt
        #print(wt,lt,pt)
        word_h = str(r_a[1]).split('\'')[1]
        wh = word_h.split('{')[0] #palabra tal cual
        lh = word_h.split('}')[0].split(',')[0].split('{')[1] #palabra lematizada
        ph = word_h.split('}')[0].split(',')[1] # pos de la palabra
        word = wh + "," + ph # palabra con el pos correspondiente
        #print(wh,lh,ph)
        if word not in palabras_encontradas:
            #Primero preguntamos si la palabra se encuentra en el grafo de la hipÃ³tesis
            if word in GH:
                # la palabra se encuentra en el grafo del texto
                if word in G:
                    print("El token: ",word," se encuentra en el Texto")
                    preds=GH.predecessors(word)
                    for e in preds:
                        po=e.split(",")
                        if(po[-1]=="ADJ" or po[-1]=="ADV"):
                            relacion.append((e,word))
                    existentes.append((word,word,1.0))
                elif wh+",H" in G: # Esto es para validar que exista la relaciÃ³n de hiperonimia
                    #print("Hiperonimo",wh)
                    su=G.successors(wh+",H") # el vinculo se encuentra en el texto
                    for e in su:
                        if e == w_in_text:
                            existentes.append((e,wh+",H",1.0))
                            preds=GH.predecessors(word)
                            for ei in preds:
                                po=ei.split(",")
                                if(po[-1]=="ADJ" or po[-1]=="ADV"):
                                    relacion.append((ei,word))
                            palabras_encontradas.append(word)
                            print("palabras encontradas",palabras_encontradas)
                            break
                elif wh+",S" in G:
                    print("Sinonimo",wh)
                    su=G.successors(wh+",S")
                    for e in su:
                        existentes.append((e,wh+",S",1.0))
                        preds=GH.predecessors(word)
                        for ei in preds:
                            po=ei.split(",")
                            if(po[-1]=="ADJ" or po[-1]=="ADV"):
                                relacion.append((e,e))
                        break
                else: ## buscamos relaciones en el grafo de wordnet
                    if pt == "NOUN" or pt=="PRON":
                        w1_=wn.synsets(wt,wn.NOUN)
                    elif pt == "VERB":
                        w1_=wn.synsets(wt,wn.VERB)
                    elif pt == "ADJ":
                        w1_=wn.synsets(wt,wn.ADJ)
                    elif pt == "ADV":
                        w1_=wn.synsets(wt,wn.ADV)
                    else:
                        w1_=wn.synsets(wt)
                    if ph == "NOUN" or ph=="PRON":
                        w2_=wn.synsets(wh,wn.NOUN)
                    elif ph == "VERB":
                        w2_=wn.synsets(wh,wn.VERB)
                    elif ph == "ADJ":
                        w2_=wn.synsets(wh,wn.ADJ)
                    elif ph == "ADV":
                        w2_=wn.synsets(wh,wn.ADV)
                    else:
                        w2_=wn.synsets(wh)
                    max_r=0
                    sim=0
                    sim_h=0
                    max_lch=0
                    max_root=0
                    cam_sh=0
                    for a in w1_:
                        for b in w2_:
                            sim=a.wup_similarity(b)
                            if sim>max_r:
                                sim_h=b.path_similarity(a, simulate_root=False)
                                if sim_h!=None:
                                    max_r=sim#+sim_h
                                    max_root=sim_h
                                    palabra1=a
                                    palabra2=b
                                    cam_sh=a.shortest_path_distance(b)
                                    if pt==ph:
                                        max_lch=a.lch_similarity(b)
                                    else:
                                        max_lch=0
                                        #print(p1,p2,max_r)
                                else:
                                    sim_h=b.path_similarity(a)    
                                    if pt==ph and a.pos()==b.pos():
                                        max_lch=a.lch_similarity(b)
                                    else:
                                        max_lch=0
                                    cam_sh=a.shortest_path_distance(b)
                                    max_r=sim
                                    max_root=sim_h
                                    palabra1=a
                                    palabra2=b
                    #Guardamos relaciones que no se consideraron como ADJ, ADV y NOUN
                    susc=GH.successors(word)
                    for ei in susc:
                        po=ei.split(",")
                        if(ph=="NOUN" and (po[-1]=="ADJ" or po[-1]=="ADV")):
                            if (ei,word) not in relacion:
                                relacion.append((ei,word))                           
                    if max_r>=0.5:
                        print("Guardado: ",w_in_text,word,"wup: ",max_r,sim,"path: ",sim_h,"lch: ",max_lch,"max_root: ",max_root,"shorthes path: ",cam_sh)
                        existentes.append((w_in_text, word,max_r))
                        razonamiento.append((palabra1,"->",palabra2,palabra1.definition(),max_r))
            else:
                relacion.append((word_t,word_h))
                print("No esta en grafo H:",word)
    print()
    print("Existentes",existentes)
    print("Faltantes",faltantes)
    print("Relaciones a buscar",relacion)
    print()

    print(set(razonamiento))
    return list(set(existentes)),relacion
```
