```python
def getRelacionesMA(indice,df):
    todas_relaciones=[]
    t=pd.DataFrame(df.at[indice,"Text"][1:])
    h=pd.DataFrame(df.at[indice,"Hipotesis"][1:])
    ma=pd.DataFrame(df.at[indice,"M_Align"][1:len(t)+1,1:len(h)+1],columns=h,index=t)
    for col in ma.columns:
        indices=[]
        if 'NOUN' in str(col) or 'VERB' in str(col) or 'ADJ' in str(col) or 'PRON' in str(col):
            qu75=ma[col].quantile(0.75)
            #print(ma[col].idxmax,ma[col].max)
            for index,strings in ma.iterrows():
                #print(index)
                if 'NOUN' in str(index) or 'VERB' in str(index):
                    if strings[col]>=qu75:
                        #print(index,strings)
                        indices.append((index,col,strings[col]))
                        todas_relaciones.append((index,col))
    for index,strings in ma.iterrows():    
        if 'NOUN' in str(index) or 'VERB' in str(index) or 'PRON' in str(index):
            qu75=strings.quantile(0.75)
            #print(qu75)
            indices=[]
            for col in ma.columns:
                if 'NOUN' in str(col) or 'VERB' in str(col) or 'PRON' in str(col):
                    #print(qu75)
                    if strings[col]>=qu75:
                        #print(index,col)
                        indices.append((index,col,strings[col]))
                        todas_relaciones.append((index,col))
    return set(todas_relaciones)
```

```python
def tokenizar(texto,nlp,wn,hipotesis=False):
    sinonimos=set()
    #print(texto)
    lista=[]
    doc = nlp(texto)
    l=[]
    for token in doc:
        #print(token.text, token.pos_,token.dep_, token.head.text, token.head.pos_)
        #print(str(token.pos_))
        if str(token.pos_):
            #print(token.text,token.pos_)
            sinonimos=set()
            if(token.pos_ == 'NOUN' or token.pos_ == 'PRON'):
                w =wn.synsets(token.text,wn.NOUN)                
            elif(token.pos_ == 'VERB'):
                w =wn.synsets(token.text,wn.VERB)
            elif(token.pos_ == 'ADJ'):
                w =wn.synsets(token.text,wn.ADJ)
            elif(token.pos_ == 'ADV'):
                w =wn.synsets(token.text,wn.ADV)
            else:
                w =wn.synsets(token.text)
            #SINONIMOS
            if w==[]:
                w = w =wn.synsets(token.lemma_)
            for wi in w:
                #print(wi)
                for sin in wi.lemma_names():
                    sinonimos.add(sin)
                    #print("Checar ",sin)
                    #print("definition:",wi,wi.definition())
                    d_ = nlp(wi.definition())
                    for t in d_:
                       for child in t.children:
                           if child.pos_ in ['NOUN','ADV','ADJ','VERB','PRON'] and t.pos_ in ['NOUN','ADV','ADJ','VERB','PRON']:
                               #lista.append((str(t.text)+","+str(t.pos_),str(child.text)+","+str(child.pos_)))
                               #print((str(t.text)+","+str(t.pos_),str(child.text)+","+str(child.pos_)))
                               #if child.pos_ in ['NOUN','ADJ']:
                                   #lista.append((str(sin)+","+str(token.pos_),str(child.text)+","+str(child.pos_)))
                               lista.append((str(sin)+","+str(token.pos_),str(child.text)+",A"))

            for e in sinonimos:
                if(e!=token.text):
                    lista.append((e+",S",str(token.text)+","+str(token.pos_)))
                #lista.append((str(token.text)+",S",e))
            

            if hipotesis==False:
                            # HIPERONIMOS
                            for wi in w:
                                paths = wi.hypernym_paths()
                                #print("CAMINOS",wi,paths)
                                for p in paths:
                                    if(len(p)>1 and len(p)<4):
                                        #print("caminos",p)
                                        j=0
                                        for pi in p[::-1]:
                                            #print(pi,token.text)
                                            #print(pi.lemma_names())
                                            for pis in pi.lemma_names():
                                                lista.append((pis+",H",str(token.text)+","+str(token.pos_)))
                                                #lista.append((str(token.text)+",H",pis))
                                            #if j==0:
                                            #    primero=pi
                                            #    lista.append((str(token.text)+","+str(token.head.pos_),pi))
                                            #    j+=1
                                            #else:
                                            #    print("nodo",pi,pi.definition())
                                            #    lista.append((primero,pi))
                                            #    primero=pi
                            #for p in paths:
                                    #    for pi in p:
                                    #        print(pi,pi.lemma_names(),pi.definition())
                                    #for li in wi.lemma_names():
                                    #    lista.append((token.text,li))
                            #print(sinonimos)
                        
            for child in token.children:
                if child.pos_ not in ['PUNCT','DET']:
                    lista.append((str(token.text)+","+str(token.pos_),str(child.text)+","+str(child.pos_)))
            l=set(lista)
    return list(l)
```
