# RTE 3

```python
df_rte3=pd.read_csv("data/RTE3/RTE3_TEST.csv")
#df_rte3.apply(lambda x : str(x).strip())
df_rte3
```

```python
temp=df_rte3.copy()
for index,strings in temp.iterrows():
    limpio=str(strings["sentence1"]).strip()
    df_rte3.at[index,"sentence1"]=limpio
    limpio=str(strings["sentence2"]).strip()
    df_rte3.at[index,"sentence2"]=limpio
```

```python
df_rte3
```

```python
df_rte3=df_rte3[["sentence1","sentence2","gold_label"]]
df_rte3
```

```python
df_rte3[:100].to_csv("data/RTE3/TEST_1.csv",index=False)
df_rte3[100:200].to_csv("data/RTE3/TEST_2.csv",index=False)
df_rte3[200:300].to_csv("data/RTE3/TEST_3.csv",index=False)
df_rte3[300:400].to_csv("data/RTE3/TEST_4.csv",index=False)
df_rte3[400:500].to_csv("data/RTE3/TEST_5.csv",index=False)
df_rte3[500:600].to_csv("data/RTE3/TEST_6.csv",index=False)
df_rte3[600:700].to_csv("data/RTE3/TEST_7.csv",index=False)
df_rte3[700:].to_csv("data/RTE3/TEST_8.csv",index=False)
```
