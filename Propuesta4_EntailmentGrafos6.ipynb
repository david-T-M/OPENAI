```python
def procesar(t):
    doc = nlp(t)
    ent_rel=[]
    subtexto=""
    anterior= False
    for token in doc:    
        #print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_)
        if token.pos_ in ["NOUN","PRON","PROPN"]:
            if anterior:
                #print("verbo:", subtexto)
                if subtexto!="":
                    ent_rel.append(subtexto)
                subtexto=token.text+" "
                #print("entidad: ",token.text)
                #ent_rel.append(token.text)
                
                anterior=False
            else:
                subtexto+=token.text+" "
        elif token.pos_=="ADJ":
            if anterior:
                if subtexto!="":
                    #print("verbo:", subtexto)
                    ent_rel.append(subtexto)
                anterior=False
                subtexto=""
                subtexto+=token.text + " "
            else:
                subtexto+=token.text + " "
        elif token.pos_=="ADV" and anterior:
            subtexto+=token.text+ " "
            #print("verbo:", subtexto)
            ent_rel.append(subtexto)
            subtexto=""
            anterior=False
        elif token.pos_== "VERB":
            if subtexto!="":
                #print("entidad:",subtexto)
                ent_rel.append(subtexto)
                subtexto=""
                subtexto+=token.text+" "
            else:
                subtexto+=token.text+" "
            anterior=True
        else:
            if token.pos_ not in ["ADP","CCONJ","AUX","DET","PART"]:
                subtexto+=token.text + " "
            elif token.pos_=="PUNCT":
                subtexto+=""
            else:
                if subtexto!="":
                    #print("entidad: ",subtexto)
                    ent_rel.append(subtexto)
                    subtexto=""
    if subtexto!="":
        #print("verbo o entidad:",subtexto)
        ent_rel.append(subtexto)
    return ent_rel
```

```python
def tokenizar(texto,nlp,wn,hipotesis=False):
    sinonimos=set()
    #print(texto)
    lista=[]
    doc = nlp(texto)
    l=[]
    validados=[]
    for token in doc:
        #print(token.text, token.pos_,token.dep_, token.head.text, token.head.pos_)
        #print(str(token.pos_))
        if str(token.pos_):
            #print(token.text,token.pos_)
            sinonimos=set()
            if(token.pos_ == 'NOUN' or token.pos_ == 'PRON'):
                w =wn.synsets(token.text,wn.NOUN)                
            elif(token.pos_ == 'VERB'):
                w =wn.synsets(token.text,wn.VERB)
            elif(token.pos_ == 'ADJ'):
                w =wn.synsets(token.text,wn.ADJ)
            elif(token.pos_ == 'ADV'):
                w =wn.synsets(token.text,wn.ADV)
            else:
                w =wn.synsets(token.text)
            if token.text in validados:
                i=lista.count(token.text)
                for child in token.children:
                    if child.pos_ not in ['PUNCT','DET','AUX','CCONJ']:
                        lista.append((str(token.text+str(i+1))+","+str(token.pos_),str(child.text)+","+str(child.pos_)))
                        print(token.text, token.pos_,token.dep_, token.head.text, token.head.pos_)
            else:
                for child in token.children:
                    if child.pos_ not in ['PUNCT','DET','AUX','CCONJ']:
                        lista.append((str(token.text)+","+str(token.pos_),str(child.text)+","+str(child.pos_)))
                        print(token.text, token.pos_,token.dep_, token.head.text, token.head.pos_)
            l=set(lista)
            validados.append(token.text)
    return list(l)
```

```python
def revisar_subject(st,sh,wn):
    st_s=st.split()
    sh_s=sh.split()
    for i in st_s:
        for j in sh_s:
            w1 = wn.synsets(i,wn.NOUN)
            w2 = wn.synsets(j,wn.NOUN) # n denotes noun
            for a in w1:
                for b in w2:
                    #print(a,b,a.lowest_common_hypernyms(b))
                    sim=a.wup_similarity(b)
                    if sim>=0.8:
                        for e in a.examples():
                            print(e)
                            print(sim,a.definition(),b.definition(),a.examples())
                            return sim
    return 0.0
```
